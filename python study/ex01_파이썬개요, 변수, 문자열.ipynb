```python
x = 100
y = 200
add = x + y
u1 = f"{x} {y}  {add}."
u2 = "{} {}  {}.".format(x,y,add)
u3 = "%d %d  %d."%(x,y,add)
print(u1)
print(u2)
print(u3)
```
